## Data PreProcessing

```python
df['variety'].value_counts()
```

```python
df.isnull().any()
```

#### Label Encoding

##### To label the variety of flower as 0,1,2

```python
labelEncoder = LabelEncoder()

df['variety'] = labelEncoder.fit_transform(df['variety'])

df['variety'].unique()
```

```python
y = df['variety']
X = df.drop(columns=['variety'], axis=1)
X.head()
```

```python
y.head()
```

```python
X = df.iloc[:, [2,3]].values
y = df.iloc[:, 4].values
y
```

### Look for correlated variables

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(), annot=True, cmap='cubehelix_r' )
plt.show()
```

# Model design

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
##Accuracy

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt

print('The accuracy of DecisionTreeClassifier is : %.2f'%accuracy_score(y_pred,y_test))
```
