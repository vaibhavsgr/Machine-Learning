```python
import pandas as pd
import nltk 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data.csv', encoding = "ISO-8859-1")
df.shape
```

```python
train = df[df['Date']<'20150101']
test = df[df['Date']>'20141231']
```

```python
train.shape
```

```python
test.shape
```

```python
#Feature engineering - removing anything but english alphabets from headlines

data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]", " ", inplace=True, regex=True)
data.head()
```

```python
data.iloc[:1,:27]
```

```python
data.columns
```

```python
new_index = [str(i) for i in range(1,26)]
data.columns = new_index
data.head()
```

```python
for i in new_index:
    data[i] = data[i].str.lower()
```

```python
data.iloc[1]
```

```python
headlines=[]
for row in range(0, len(data.index)):
    headlines.append(" ".join(str(h) for h in data.iloc[row,0:25]))
data.shape
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#countvector = CountVectorizer(ngram_range=(2,2))
countvector = TfidfVectorizer()
traindataset = countvector.fit_transform(headlines)
traindataset.shape
```

```python
RF = RandomForestClassifier(n_estimators=200,criterion='entropy')
RF.fit(traindataset, train['Label'])
X = countvector.fit_transform(headlines).toarray()
```

```python
X
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(h) for h in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = RF.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
